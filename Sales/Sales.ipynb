# Remove Duplicate Records

```python
# Check duplicate rows
df.duplicated().sum()

# Remove duplicates
df.drop_duplicates(inplace=True)
```

```python
# Create time-based features
df['YEAR'] = df['ORDERDATE'].dt.year
df['MONTH'] = df['ORDERDATE'].dt.month
df['QUARTER'] = df['ORDERDATE'].dt.quarter
```

```python
# Calculate estimated profit
df['ESTIMATED_PROFIT'] = (df['PRICEEACH'] - df['MSRP']) * df['QUANTITYORDERED']
```

* Key Performance Indicators (KPIs)

```python
# Total Sales
total_sales = df['SALES'].sum()

# Total Estimated Profit
total_profit = df['ESTIMATED_PROFIT'].sum()

# Average Order Value
average_order_value = df['SALES'].mean()

total_sales, total_profit, average_order_value
```

* Product-Level Analysis

```python
# Sales and profit by product line
product_analysis = df.groupby('PRODUCTLINE')[['SALES', 'ESTIMATED_PROFIT']].sum().sort_values(by='SALES', ascending=False)
product_analysis
```

* Country-Level Sales Analysis

```python
# Total sales by country
country_sales = df.groupby('COUNTRY')['SALES'].sum().sort_values(ascending=False)
country_sales
```

* Customer-Level Analysis

```python
# Customer summary
customer_analysis = df.groupby('CUSTOMERNAME').agg({
    'SALES': 'sum',
    'ORDERNUMBER': 'nunique'
}).rename(columns={'ORDERNUMBER': 'TOTAL_ORDERS'})

# Top 10 customers by sales
customer_analysis.sort_values(by='SALES', ascending=False).head(10)
```

* Trend Analysis (Year-wise Sales)

```python
# Year-wise sales trend
yearly_sales = df.groupby('YEAR')['SALES'].sum()
yearly_sales
```

# TERRITORY-WISE SALES ANALYSIS

* Which territory generates the highest revenue?

```python
# Total sales by territory
territory_sales = (
    df.groupby('TERRITORY')['SALES']
    .sum()
    .sort_values(ascending=False)
)

territory_sales
```

 * Chart: Territory-wise Total Sales (Bar Chart)

```python
# Bar chart for total sales by territory
territory_sales.plot(kind='bar', figsize=(8,5))

plt.title("Territory-wise Total Sales")
plt.xlabel("Territory")
plt.ylabel("Total Sales")
plt.grid(True)
plt.show()
```

* Average Sales per Order by Territory (Order Value)
  * Which territory has higher-value orders on average?

```python
# Average sales per order by territory
avg_territory_sales = (
    df.groupby('TERRITORY')['SALES']
    .mean()
    .sort_values(ascending=False)
)

avg_territory_sales
```

* Chart: Average Sales per Order (Horizontal Bar)

```python
# Horizontal bar chart for average sales
avg_territory_sales.plot(kind='barh', figsize=(8,5))

plt.title("Average Sales per Order by Territory")
plt.xlabel("Average Sales")
plt.ylabel("Territory")
plt.grid(True)
plt.show()
```

* Territory Contribution Percentage (Executive Insight)

   * How much does each territory contribute to total company sales?

```python
# Territory contribution percentage
territory_contribution = (
    df.groupby('TERRITORY')['SALES'].sum() /
    df['SALES'].sum()
) * 100

territory_contribution = territory_contribution.sort_values(ascending=False)

territory_contribution
```

* Chart: Sales Contribution by Territory (Pie Chart)

```python
# Pie chart for territory contribution
territory_contribution.plot(
    kind='pie',
    autopct='%1.1f%%',
    figsize=(6,6)
)

plt.title("Sales Contribution by Territory")
plt.ylabel("")
plt.show()
```

* Territory-wise Sales Volume (Units Sold)

    * Which territory sells the highest quantity of products?

```python
# Total quantity ordered by territory
territory_quantity = (
    df.groupby('TERRITORY')['QUANTITYORDERED']
    .sum()
    .sort_values(ascending=False)
)

territory_quantity
```

```python
# Horizontal bar chart for total quantity ordered by territory
territory_quantity.plot(kind='barh', figsize=(8,5))

plt.title("Total Quantity Ordered by Territory")
plt.xlabel("Total Quantity Ordered")
plt.ylabel("Territory")
plt.grid(True)
plt.show()
```
